8. Write the code to find the percentage missing value (column-wise) in master_dataframe. Drop
unnecessary columns.

```python
for col in master_dataframe:
    total_nan_in_col = master_dataframe[col].isnull().sum()
    percent_nan_in_col = total_nan_in_col/no_rows * 100
    print(col, percent_nan_in_col, "%")
```

12. Create the ‘main_category’ column by extracting the main sector using the column
‘category_list’.

```python
master_dataframe['category_list'] = master_dataframe['category_list'].astype('string')
def getMainCategory(x):
    if isinstance(x, str):
        return x.split("|")[0]
    else: return 'nan'

main_category = []
for x in master_dataframe['category_list']:
    main_category.append(getMainCategory(x))
master_dataframe['main_category'] = main_category
```

13. Using the user defined function convert the ‘rasied_amount_used’ column into a categorical
column as followa)
If the amount is less than 5 Millions then recode as TypeA.
b) If the amount is greater than 5 Millions and less than 7 Millions then recode as TypeB.
c) If the amount is greater than 7 Millions then recode as TypeC.

```python
def getCategory(x):
    if x < 5000000:
        return "TypeA"
    elif x >= 5000000 and x < 7000000:
        return "TypeB"
    else: return "TypeC"
#     match x:
#         case x < 5000000:
#             return "TypeA"
#         case x >= 5000000 and x < 7000000:
#             return "TypeB"
#         case x >= 7000000:
#             return "TypeC"
master_dataframe['raised_amount_usd'] = master_dataframe['raised_amount_usd'].apply(lambda x: getCategory(x))
```
